```python
def power(x):
    return x * x
def under_3(x):
    return x < 3

x_list = [1,2,3,4,5]
#  map(함수, 리스트)함수를 이용하여  x_list의 각 요소에 power() 함수 실행 
# output_a = map(power, x_list)   # power()  함수 호출 -> 리스트의 각 요소별로 호출
output_a = [ x * x for x in x_list ]
print("map() 함수 실행 결과 ",  list(output_a))

# filter( 함수, 리스트)함수를 이용하여  x_list의 각 요소에 under_3() 함수 실행 
# output_b = filter(under_3, x_list)   # power()  함수 호출 -> 리스트의 각 요소별로 호출
output_b = [ x for x in x_list if x < 3 ]
print("filter() 함수 실행 결과 ",  list(output_b))
```

```python
# info.txt 파일 생성 : 1000명의 이름, 몸무게, 키 자료를 저장
import random
names = list('가나다라마바사아자차카타파하')

# 파일을 쓰기 모드로 open
with open('info.txt', 'w') as file:
    for _ in range(1000):
        # names에서 임의로 2글자 선택
        name = random.choice(names) + random.choice(names)
        weight = random.randrange(40,100)
        height = random.randrange(140,200)
        
        file.write(f'{name}, {weight}, {height}\n')
```

```python
# 저장된 자료를 읽어와서 이름, 몸무게, 키, BMI 출력하면서 정상 체중 or 과체중 or 저체중 확인
# BMI = 몸무게/(키**2)
# 저체중 : 20 미만, 정상 : 20 - 24, 과체중 : 25-29, 비만 : 30 이상
with open('info.txt', 'r') as file:
    for line in file:
        name, weight, height = line.split(', ')
        height = height[:-1]  # 마지막 글자 '\n' 제거
        bmi = int(int(weight) / ((int(height)/100)**2))
        if bmi < 20:
            result = '저체중'
        elif bmi <= 24:
            result = '정상 체중'
        elif bmi <= 29:
            result = '과체중'
        else:
            result = '비만'
        print(f'이름:{name}  체중:{weight}  키:{height}  BMI:{bmi}  {result}')
```

## chapter 6-1 구문 오류와 예외
- try ~ except ~ else ~ finally 구문

```python
try:  # 예외가 발생할 가능성이 있는 코드
    num = int(input('숫자 입력 : '))
except:  # 예외가 발생했을 때 실행할 코드
    print('정수 입력 : ')
else:  # 예외가 발생하지 않았을 때 실행할 코드
    print('원의 반지름 : ', num)
    print('원의 둘레 : ', num*3.14*2)
    print('원의 넓이 : ', num*num*3.14)
finally:  # 무조건 실행하는 코드
    print('end')
```

```python
# finally 구문 사용 예
try:
    file = open('info.txt', 'r')
    file.read()
    file.write('test')
except Exception as e:
    print(e)
finally:
    print('finally -- ')
    file.close()
```

```python
# 함수 초기 리턴시에도 finally는 무조건 실행
def test():
    print('test func start --- ')
    try:
        print('try 구문 실행')
        return
        print('try return 실행')
    except:
        print('except 구문 실행')
    finally:
        print('finally 실행')  # 무조건 실행
    print('test func end --- ')
```

```python
test()
```

```python
# 반복문에서 사용
while True:
    try:
        print('try 구문 실행')
        break
        print('try return 실행')
    except:
        print('except 구문 실행')
    finally:
        print('finally 실행')  # 무조건 실행
    print('while end --- ')
print('program end')        
```

## chapter 6-2 예외 고급

```python
# 변수를 선언합니다.
list_number = [52, 273, 32, 72, 100]

# try except 구문으로 예외를 처리합니다.
try:
    # 숫자를 입력 받습니다.
    number_input = int(input("정수 입력> "))
    # 리스트의 요소를 출력합니다.
    print("{}번째 요소: {}".format(number_input, list_number[number_input]))
#     예외.발생해주세요()
except ValueError as exception:
    # ValueError가 발생하는 경우
    print("정수를 입력해 주세요!")
    print(type(exception), exception)
except IndexError as exception:
    # IndexError가 발생하는 경우
    print("리스트의 인덱스를 벗어났어요!")
    print(type(exception), exception)
except Exception as exception:
    # 이외의 예외가 발생한 경우
    print("미리 파악하지 못한 예외가 발생했습니다.")
    print(type(exception), exception)
```
