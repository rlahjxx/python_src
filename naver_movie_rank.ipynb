```python
# 네이버 영화 현재 상영작에 대한 순위 이름 평점 출력
from urllib import request
from bs4 import BeautifulSoup
```

```python
# 네이버 영화 랭킹 검색 사이트 : 
url = 'https://movie.naver.com/movie/sdb/rank/rmovie.naver?sel=pnt&date=20220816'
target = request.urlopen(url)
soup = BeautifulSoup(target)
```

```python
# 영화 제목, 영화 평점을 검색해서 가져옴 -> tag를 검색해서 원하는 정보를 가져옴
movie_title = []
movie_point = []

for line in soup.find_all('tr'):  # <tr> tag 모두 검색
    title = line.find('div', class_='tit5')  # <div class='tit5' .. > text </div> 검색
#     print(title)
    if title:
#         print(title.get_text())
#         break
        movie_title.append(title.get_text().strip('\n'))
    point = line.find('td', class_='point')
    if point:
        movie_point.append(point.get_text())
print(movie_title)
print(movie_point)
for i, (title, point) in enumerate(zip(movie_title, movie_point)):
    print(f'{i+1} -> {title} : {point}')
```
